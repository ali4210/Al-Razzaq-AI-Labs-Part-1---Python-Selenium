```python
from selenium import webdriver

driver = webdriver.Chrome()
driver.get('https://example.com')  # Replace with the URL you wish to visit

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

element = driver.find_element_by_id('target-element')  # Replace with the actual element ID

driver.execute_script("arguments[0].scrollIntoView();", element)

assert element.is_displayed(), "Element is not visible on the page."
```
