```python
from selenium import webdriver

driver = webdriver.Chrome()  # Replace with `webdriver.Firefox()` if using Firefox
driver.get('file:///path/to/your/test_page.html')  # Change the path accordingly
```

```python
file_input = driver.find_element_by_id('fileInput')
```

```python
file_input.send_keys('/path/to/your/file.txt')  # Use the absolute path of your file
```

```python
submit_button = driver.find_element_by_xpath('//button[@type="submit"]')
submit_button.click()
```
