```python
from selenium import webdriver

# Set up the WebDriver
driver = webdriver.Chrome()

# Navigate to a webpage
driver.get('http://example.com')

# Set implicit wait
driver.implicitly_wait(5)

# Locate an element (the WebDriver will wait up to 5 seconds)
element = driver.find_element_by_id('someId')

# Continue your testing...
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up the WebDriver
driver = webdriver.Chrome()

# Navigate to a webpage
driver.get('http://example.com')

# Use WebDriverWait to wait for element to be clickable
wait = WebDriverWait(driver, 10)
element = wait.until(
    EC.element_to_be_clickable((By.ID, 'someClickableElementId'))
)

# Perform actions on the element
element.click()

# Continue your testing...
driver.quit()
```
