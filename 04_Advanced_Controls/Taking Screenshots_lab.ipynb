```python
from selenium import webdriver

# Initialize the WebDriver instance
driver = webdriver.Chrome()  # Use webdriver.Firefox() for Firefox
driver.get('https://www.example.com')

# Define the filename for the screenshot
screenshot_file = "screenshot.png"

# Capture the screenshot
driver.save_screenshot(screenshot_file)

import time

# Generate a timestamp string
timestamp = time.strftime("%Y%m%d-%H%M%S")

# Update filename to include timestamp
timestamped_file = f"screenshot_{timestamp}.png"

# Save the screenshot with the new filename
driver.save_screenshot(timestamped_file)

driver.quit()
```
