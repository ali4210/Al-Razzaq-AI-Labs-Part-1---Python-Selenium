```python
from selenium import webdriver

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the test web page
driver.get("http://the-internet.herokuapp.com/javascript_alerts")
```

```python
# Find the element that triggers the alert and click it
alert_button = driver.find_element_by_xpath("//button[text()='Click for JS Alert']")
alert_button.click()
```

```python
# Switch to alert
alert = driver.switch_to.alert
```

```python
# Accept the alert
alert.accept()
print("Alert accepted successfully.")
```

```python
# Dismiss the alert if needed
# alert.dismiss()
```

```python
# Get text from alert
alert_text = alert.text
print(f"Alert says: {alert_text}")
```

```python
# Close the browser
driver.quit()
```
