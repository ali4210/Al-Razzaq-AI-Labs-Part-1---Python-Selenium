```python
// Using JavaScript
const button = document.getElementById('sampleButton');
```

```python
button.click();
```

```python
from selenium import webdriver

# Initialize the browser driver
driver = webdriver.Chrome()

# Open the webpage
driver.get('file://path-to-your-sample-page.html')

# Locate the button and click
button = driver.find_element_by_id('sampleButton')
button.click()

# Close the browser
driver.quit()
```

```python
const puppeteer = require('puppeteer');

(async () => {
  const browser = await puppeteer.launch();
  const page = await browser.newPage();
  await page.goto('file://path-to-your-sample-page.html');
  await page.click('#sampleButton');
  await browser.close();
})();
```

```python
button.addEventListener('click', () => {
  document.body.innerHTML += '<p>Button clicked!</p>';
});
```

```python
# After clicking, check the current URL
assert driver.current_url == 'http://expected-url-after-click.com'
```

```python
# Verify if the new element is present
assert "Button clicked!" in driver.page_source
```
