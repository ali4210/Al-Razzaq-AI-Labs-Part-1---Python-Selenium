```python
from selenium import webdriver

# Instantiate a browser driver
driver = webdriver.Chrome()

# Open a webpage
driver.get("http://example.com/login")

# Locate the text input field by its ID
input_field = driver.find_element_by_id("username")
```

```python
input_field.send_keys("testuser")
```

```python
<input type="password" id="password" name="password" />
```

```python
# Locate the password input field by its ID
password_field = driver.find_element_by_id("password")
# Enter a password into the password field
password_field.send_keys("securepassword123")
```

```python
print("Username and password fields populated successfully.")
```

```python
driver.quit()
```
